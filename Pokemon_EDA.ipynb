```python
#Loading the required libraries
import numpy as np
import matplotlib as mlt
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import statsmodels.api as sd
from sklearn import tree
import sklearn.metrics as sm
import warnings
from IPython.display import Image, HTML
warnings.filterwarnings("ignore")

```

```python
import pandas as pd
file_url = "https://raw.githubusercontent.com/Payal2000/Pokemon-Analysis/main/pokemon.csv"
df = pd.read_csv(file_url)
```

```python
##Reading first 5 rows
df.head(5)
```

```python
df2 = df[['pokemon_name','pic_url']]
df2['pic_url'] = df2['pic_url'].apply(lambda x: f"<img src='" + x +"' width='200px'>")
```

```python
from IPython.core.display import display, HTML

# Number of columns for the grid
num_columns = 5

# Reshape the dataframe
reshaped_df = df2.groupby(df2.index // num_columns)

# Create the grid HTML
html_str = '<div style="display: grid; grid-template-columns: repeat({0}, 1fr); gap: 10px;">'.format(num_columns)

for name, group in reshaped_df:
    for index, row in group.iterrows():
        html_str += '''
        <div style="text-align: center; border: 1px solid black; padding: 10px;">
            <div><strong>Index:</strong> {0}</div>
            <div>{1}</div>
            {2}
        </div>
        '''.format(index, row['pokemon_name'], row['pic_url'])

html_str += '</div>'

# Display the grid with grid lines
display(HTML(html_str))
```
